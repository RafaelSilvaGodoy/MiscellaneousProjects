```python
import numpy as np
from scipy.integrate import quad
```

```python
def gamma(x,t):
    return (np.exp(-x)*x**(t-1))

def factorial(N):
  return quad(gamma, 0, np.inf, args=(N+1))[0]

def beta(x,y):
  return (factorial(x)*factorial(y))/factorial(x+y)
```

# Gamma Function

The gamma function is the function that calculates the factorial of any complex number.

The function operates like: $\Gamma$(n) = (n-1)!

For more informations check: https://en.wikipedia.org/wiki/Gamma_function

image.png

```python
X = np.linspace(-4, 5, num=(5-(-4)+1)*4)

y = []
for x in X:
  y.append(factorial(x))
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[go.Scatter(x=X, y=y, line=dict(color="crimson"))],
    layout=dict(title=dict(text="Factorial Graph"))
)
fig.update_layout(yaxis_range=[-30,25])
fig.show()
```

# Beta Function

The beta function is useful in computing and representing the scattering amplitude, it is closely tied with the gamma function and plays an important role in calculus.

For more information check: https://en.wikipedia.org/wiki/Beta_function

image.png

```python
import random

X = np.random.random_sample(size = 400)*6
Y = np.random.random_sample(size = 400)*6

z = []
for x,y in zip(X,Y):
  z.append(beta(x,y))
```

```python

fig = go.Figure(data=[go.Scatter3d(x=X, y=Y, z=z,
                                   mode='markers',
                                   marker=dict(
                                                size=12,
                                                color=z,                
                                                colorscale='Viridis',   
                                                opacity=0.8
                                            )
                                   )]
                )
fig.update_layout(title_text="Beta Function")
fig.show()
```
